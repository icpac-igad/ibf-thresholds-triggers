```python
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
import cartopy.crs as ccrs
```

```python
load_dotenv()

data_path=os.getenv("data_path")
```

```python


shapefile_path = f'{data_path}ea_ghcf_icpac.shp'
gdf = gpd.read_file(shapefile_path)

# Select the time slices for comparison
#kn_seas51_tprate=xr.open_dataset(f'{input_path}ken_seas51-tprate.nc')

#obs2_left = kn_seas51_tprate.isel(number=1,forecastMonth=1,time=0)
kn_chd=xr.open_dataset(f'{data_path}ea_seas51_tprate_1981_2023.nc')
ds_l2=xr.open_dataset(f'{data_path}ea_seas51_spi3_xclim_20240306.nc')


# Create a figure with two subplots
# Create a figure with four subplots
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5), subplot_kw={'projection': ccrs.PlateCarree()})

# Plot the left data on the first subplot
ol1=kn_chd.isel(time=0,number=1,forecastMonth=1)
ol1['tprate'].plot(ax=axes[0], transform=ccrs.PlateCarree(), cmap='coolwarm')
#axes[0].set_extent([31, 43, -6, 8])
gdf.plot(ax=axes[0], facecolor='none', edgecolor='black')
axes[0].gridlines(draw_labels=True)
datet_left = ol1['time'].values
axes[0].set_title(f'Left Map: Value on {datet_left}')

ol2=kn_chd.isel(time=1,number=1,forecastMonth=1)
ol2['tprate'].plot(ax=axes[1], transform=ccrs.PlateCarree(), cmap='coolwarm')
#axes[1].set_extent([31, 43, -6, 8])
gdf.plot(ax=axes[1], facecolor='none', edgecolor='black')
axes[1].gridlines(draw_labels=True)
datet_left = ol2['time'].values
axes[1].set_title(f'Left Map: Value on {datet_left}')

ol3=kn_chd.isel(time=2,number=1,forecastMonth=1)
ol3['tprate'].plot(ax=axes[2], transform=ccrs.PlateCarree(), cmap='coolwarm')
#axes[2].set_extent([31, 43, -6, 8])
gdf.plot(ax=axes[2], facecolor='none', edgecolor='black')
axes[2].gridlines(draw_labels=True)
datet_left = ol3['time'].values
axes[2].set_title(f'Left Map: Value on {datet_left}')


# Plot the right data on the second subplot
ds_l3=ds_l2.isel(time=2,member=1,lead=1)
ds_l3['spi3'].plot(ax=axes[3], transform=ccrs.PlateCarree(), cmap='coolwarm')
#axes[3].set_extent([31, 43, -6, 8])
gdf.plot(ax=axes[3], facecolor='none', edgecolor='black')
axes[3].gridlines(draw_labels=True)
datet_right = ds_l3['time'].values
axes[3].set_title(f'Right Map: Value on {datet_right}')


# Adjust layout
plt.tight_layout()

plt.show()
```
